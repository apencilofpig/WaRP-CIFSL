```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import random
from matplotlib.colors import CSS4_COLORS
import colorsys
```

```python
test_features_tsne = np.load('t-sne/resnet18/alpha0.3/features_tsne.npy')
label_list = np.load('t-sne/resnet18/alpha0.3/label_list.npy')
```

```python
def generate_colors(num_colors):
    hues = [(i / num_colors) for i in range(num_colors)]
    saturation = 0.6
    lightness = 0.5
    colors = [colorsys.hls_to_rgb(h, lightness, saturation) for h in hues]
    return colors
```

```python
# color_names = list(CSS4_COLORS.keys())
# random_colors = random.sample(color_names, 36)
random.seed(1)
random_colors = generate_colors(36)
random.shuffle(random_colors)
```

```python
# 
for label in np.unique(label_list):
    if label >= 26:
        continue
    indices = label_list == label
    plt.scatter(test_features_tsne[indices, 0], test_features_tsne[indices, 1], s=5, color=random_colors[label], label=f'Label {label}')
    xtext, ytext = np.median(test_features_tsne[indices, :], axis=0)
    txt = plt.text(xtext, ytext, str(label), fontsize=7)

plt.title('t-SNE Visualization of Test Data Features')
# plt.xlabel('t-SNE Component 1')
# plt.ylabel('t-SNE Component 2')
# plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.legend()
plt.savefig('t-SNE.png', dpi=700, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import colorsys

def generate_colors(num_colors):
    hues = [(i / num_colors) for i in range(num_colors)]
    saturation = 0.6
    lightness = 0.4
    colors = [colorsys.hls_to_rgb(h, lightness, saturation) for h in hues]
    return colors

# 36
num_colors = 36
colors = generate_colors(num_colors)

# 
fig, ax = plt.subplots(figsize=(12, 1))
for i, color in enumerate(colors):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
ax.set_xlim(0, num_colors)
ax.set_ylim(0, 1)
ax.axis('off')
plt.show()
```
